## model scores

```LGBMRegressor  => mean over all CV loop R2s: 0.268```

Zum Vergleich (08.04.): ```LGBMRegressor  => mean  0.230 (train: 0.425)```

Values comparable? Besides **CV loop R2**, there is also **Running mean R2** and **Running shuffle mean R2**

````python
import pickle
import pandas as pd
  
# Open the file in binary mode
with open('results_audit.pickle', 'rb') as fo:
      
    # Call load method to deserialze
    results = pickle.load(fo)
    
#print("results keys:",results.keys(),"\n")

df_scores = pd.DataFrame(results["scores"]).apply(lambda row: pd.Series({"score_1":row[0][0],"score_2":row[1][0],"curr_loop_R2":row[2][0]}),axis=1)
print(df_scores.round(3))
print("\nmeans:")
print(df_scores.mean())
````
